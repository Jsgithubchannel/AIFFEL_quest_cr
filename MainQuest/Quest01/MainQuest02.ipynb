```python
# Character í´ë˜ìŠ¤
# ì†ì„± - ì´ë¦„, ë ˆë²¨, ì²´ë ¥, ê³µê²©ë ¥, ë°©ì–´ë ¥
# ë©”ì„œë“œ - is_alive: í˜„ì¬ ì²´ë ¥ì´ 0 ì´ìƒì¸ì§€ bool ê°’ ë°˜í™˜
# ë©”ì„œë“œ - take_damage: ë°›ì€ ë°ë¯¸ì§€ - ë³¸ì¸ì˜ ë°©ì–´ë ¥ë§Œí¼ í˜„ì¬ ì²´ë ¥ ê°ì†Œ
# ë©”ì„œë“œ - attack_targe: íƒ€ê²Ÿì—ê²Œ ë°ë¯¸ì§€ë¥¼ ì…í˜, ë°ë¯¸ì§€ëŠ” 1~ê³µê²©ë ¥ ì‚¬ì´ ëœë¤ ì •ìˆ˜
import random as r

class Character():
    def __init__(self, name='', level=0, hp=0, atk=0, defense=0):
        self.name = name
        self.level = level
        self.hp = hp
        self.atk = atk
        self.defense = defense

    def is_alive(self):
        if self.hp >= 0:
            return True
        else:
            return False

    def take_damage(self, damage):
        actual_damage = max(0, damage - self.defense) # ë°©ì–´ë ¥ì´ ë°ë¯¸ì§€ë³´ë‹¤ í¬ë‹¤ë©´ ì²´ë ¥ ê°ì†Œ x
        self.hp -= actual_damage

    def attack_target(self, target):
        # íƒ€ê²Ÿì—ê²Œ ë°ë¯¸ì§€ë¥¼ ì…í˜, ë°ë¯¸ì§€ëŠ” 1~ê³µê²©ë ¥ ì‚¬ì´ ëœë¤ ì •ìˆ˜
        damage = r.randint(1, self.atk)
        target.take_damage(damage)
        return damage


# Player í´ë˜ìŠ¤
# Character ìƒì†
# ë ˆë²¨, ì²´ë ¥, ê³µê²©ë ¥, ë°©ì–´ë ¥ ì´ˆê¸°í™” ë° ê²½í—˜ì¹˜ ì†ì„± ì¶”ê°€
# ë©”ì„œë“œ - gain_experience: ì¸ìˆ˜ë¡œ ë°›ì€ ì •ìˆ˜ ë§Œí¼ ê²½í—˜ì¹˜ë¥¼ íšë“
# ë©”ì„œë“œ - level_up: í˜„ì¬ ê²½í—˜ì¹˜ê°€ 50ì´ìƒì´ë©´ ë ˆë²¨ì„ 1, ê³µê²©ë ¥ì„ 10, ë°©ì–´ë ¥ì„ 5ì”© up
class Player(Character):
    def __init__(self, name):
        super().__init__(name, level=1, hp=100, atk=25, defense=5)
        self.exp = 0

    def gain_experience(self, num):
        self.exp += num

    def level_up(self):
        if self.exp >= 50:
            self.level += 1
            self.atk += 10
            self.defense += 5
            self.exp -= 50  # ê²½í—˜ì¹˜ 50 ì°¨ê°
            print('-'*20, 'ë ˆë²¨ ì—…! â¬†ï¸ ğŸ”¥ğŸ”¥', '-'*20)

# Monster í´ë˜ìŠ¤
# Character ìƒì†
# ëª¬ìŠ¤í„°ì˜ ë ˆë²¨ì— ë¹„ë¡€í•˜ëŠ” ì²´ë ¥, ê³µê²©ë ¥, ë°©ì–´ë ¥ ì´ˆê¸°í™”
# ì²´ë ¥: 10~30 ì‚¬ì´ì˜ ëœë¤í•œ ì •ìˆ˜ * ë ˆë²¨
# ê³µê²©ë ¥: 5~20 ì‚¬ì´ì˜ ëœë¤í•œ ì •ìˆ˜ * ë ˆë²¨
# ë°©ì–´ë ¥: 1~5 ì‚¬ì´ì˜ ëœë¤í•œ ì •ìˆ˜ * ë ˆë²¨
class Monster(Character):
    def __init__(self, name, level):
        self.name = name
        self.level = level
        self.hp = r.randint(10,30) * self.level
        self.atk = r.randint(5,20) * self.level
        self.defense = r.randint(1,5) * self.level

# battle í•¨ìˆ˜
# Player, Monster ë‘˜ ì¤‘ í•˜ë‚˜ì˜ ì²´ë ¥ì´ 0 ì´í•˜ê°€ ë  ë•Œê¹Œì§€ ê³µê²©
# Player ìƒì¡´ -> ëª¬ìŠ¤í„° ë ˆë²¨ * 20 ê²½í—˜ì¹˜ ì¶”ê°€, ë ˆë²¨ì—…, ì „íˆ¬ ìŠ¹ë¦¬ ì¶œë ¥
# Player íŒ¨ë°° -> ì „íˆ¬ íŒ¨ë°° ì¶œë ¥
def battle(player, monsters):
    for monster in monsters:
        print(f'{monster.name}ê³¼ì˜ ì „íˆ¬ë¥¼ ì‹œì‘í•©ë‹ˆë‹¤.')
        while player.is_alive() and monster.is_alive():
            # í”Œë ˆì´ì–´ì˜ ê³µê²©
            damage = player.attack_target(monster)
            print(f'{player.name}ì´ {monster.name}ì—ê²Œ {damage}ë§Œí¼ ê³µê²©í–ˆë‹¤...!\n')
            print(f'{monster.name}ì˜ ì²´ë ¥: {monster.hp}')

            if not monster.is_alive():  # ëª¬ìŠ¤í„° ì²´ë ¥ 0 ì´í•˜ì¸ ê²½ìš°
                break

            # ëª¬ìŠ¤í„°ì˜ ê³µê²©
            if monster.is_alive():
                damage = monster.attack_target(player)
                print(f'{monster.name}ì´ {player.name}ì—ê²Œ {damage}ë§Œí¼ ê³µê²©í–ˆë‹¤...!\n')
                print(f'{player.name}ì˜ ì²´ë ¥: {player.hp}')

            if not player.is_alive():  # í”Œë ˆì´ì–´ ì²´ë ¥ 0 ì´í•˜ì¸ ê²½ìš°
                break

        # ì „íˆ¬ ì¢…ë£Œ
        if player.is_alive():
            print('-'*20)
            print('ì „íˆ¬ ìŠ¹ë¦¬!')
            print(f'{monster.name}ì„ ì´ê²¼ë‹¤! ğŸ¥¹')
            print('-'*20)
            player.gain_experience(monster.level * 20)
            player.level_up()
            continue
        else:
            print('ì „íˆ¬ íŒ¨ë°°.. ğŸ˜±')
            print('-'*20, 'ê²Œì„ ì˜¤ë²„', '-'*20)
            return


# ì‚¬ìš©ì ì´ë¦„ ë°›ì•„ì„œ Player ì¸ìŠ¤í„´ìŠ¤ ìƒì„±
# ëª¬ìŠ¤í„° ë”•ì…”ë„ˆë¦¬ë¡œ Monster ì¸ìŠ¤í„´ìŠ¤ ìƒì„±
# playerëŠ” ëª¬ìŠ¤í„° 3ë§ˆë¦¬ ì „ë¶€ ì „íˆ¬
# ì „íˆ¬ ë„ì¤‘ì— Player ì‚¬ë§ì‹œ, ê²Œì„ ì˜¤ë²„ ì¶œë ¥ í›„ ì¢…ë£Œ
monsters = []
if __name__ == '__main__':
    monster_dict = {'ìŠ¬ë¼ì„': 1, 'ê³ ë¸”ë¦°': 2, 'ì˜¤í¬': 3}
    player_name = input('í”Œë ˆì´ì–´ì˜ ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš”: ')
    player = Player(name=player_name)

    monsters = [Monster(name, level) for name,level in monster_dict.items()]
    battle(player, monsters)
```
