```python
# Character 클래스
# 속성 - 이름, 레벨, 체력, 공격력, 방어력
# 메서드 - is_alive: 현재 체력이 0 이상인지 bool 값 반환
# 메서드 - take_damage: 받은 데미지 - 본인의 방어력만큼 현재 체력 감소
# 메서드 - attack_targe: 타겟에게 데미지를 입힘, 데미지는 1~공격력 사이 랜덤 정수
import random as r

class Character():
    def __init__(self, name='', level=0, hp=0, atk=0, defense=0):
        self.name = name
        self.level = level
        self.hp = hp
        self.atk = atk
        self.defense = defense

    def is_alive(self):
        if self.hp >= 0:
            return True
        else:
            return False

    def take_damage(self, damage):
        actual_damage = max(0, damage - self.defense) # 방어력이 데미지보다 크다면 체력 감소 x
        self.hp -= actual_damage

    def attack_target(self, target):
        # 타겟에게 데미지를 입힘, 데미지는 1~공격력 사이 랜덤 정수
        damage = r.randint(1, self.atk)
        target.take_damage(damage)
        return damage


# Player 클래스
# Character 상속
# 레벨, 체력, 공격력, 방어력 초기화 및 경험치 속성 추가
# 메서드 - gain_experience: 인수로 받은 정수 만큼 경험치를 획득
# 메서드 - level_up: 현재 경험치가 50이상이면 레벨을 1, 공격력을 10, 방어력을 5씩 up
class Player(Character):
    def __init__(self, name):
        super().__init__(name, level=1, hp=100, atk=25, defense=5)
        self.exp = 0

    def gain_experience(self, num):
        self.exp += num

    def level_up(self):
        if self.exp >= 50:
            self.level += 1
            self.atk += 10
            self.defense += 5
            self.exp -= 50  # 경험치 50 차감
            print('-'*20, '레벨 업! ⬆️ 🔥🔥', '-'*20)

# Monster 클래스
# Character 상속
# 몬스터의 레벨에 비례하는 체력, 공격력, 방어력 초기화
# 체력: 10~30 사이의 랜덤한 정수 * 레벨
# 공격력: 5~20 사이의 랜덤한 정수 * 레벨
# 방어력: 1~5 사이의 랜덤한 정수 * 레벨
class Monster(Character):
    def __init__(self, name, level):
        self.name = name
        self.level = level
        self.hp = r.randint(10,30) * self.level
        self.atk = r.randint(5,20) * self.level
        self.defense = r.randint(1,5) * self.level

# battle 함수
# Player, Monster 둘 중 하나의 체력이 0 이하가 될 때까지 공격
# Player 생존 -> 몬스터 레벨 * 20 경험치 추가, 레벨업, 전투 승리 출력
# Player 패배 -> 전투 패배 출력
def battle(player, monsters):
    for monster in monsters:
        print(f'{monster.name}과의 전투를 시작합니다.')
        while player.is_alive() and monster.is_alive():
            # 플레이어의 공격
            damage = player.attack_target(monster)
            print(f'{player.name}이 {monster.name}에게 {damage}만큼 공격했다...!\n')
            print(f'{monster.name}의 체력: {monster.hp}')

            if not monster.is_alive():  # 몬스터 체력 0 이하인 경우
                break

            # 몬스터의 공격
            if monster.is_alive():
                damage = monster.attack_target(player)
                print(f'{monster.name}이 {player.name}에게 {damage}만큼 공격했다...!\n')
                print(f'{player.name}의 체력: {player.hp}')

            if not player.is_alive():  # 플레이어 체력 0 이하인 경우
                break

        # 전투 종료
        if player.is_alive():
            print('-'*20)
            print('전투 승리!')
            print(f'{monster.name}을 이겼다! 🥹')
            print('-'*20)
            player.gain_experience(monster.level * 20)
            player.level_up()
            continue
        else:
            print('전투 패배.. 😱')
            print('-'*20, '게임 오버', '-'*20)
            return


# 사용자 이름 받아서 Player 인스턴스 생성
# 몬스터 딕셔너리로 Monster 인스턴스 생성
# player는 몬스터 3마리 전부 전투
# 전투 도중에 Player 사망시, 게임 오버 출력 후 종료
monsters = []
if __name__ == '__main__':
    monster_dict = {'슬라임': 1, '고블린': 2, '오크': 3}
    player_name = input('플레이어의 이름을 입력하세요: ')
    player = Player(name=player_name)

    monsters = [Monster(name, level) for name,level in monster_dict.items()]
    battle(player, monsters)
```
