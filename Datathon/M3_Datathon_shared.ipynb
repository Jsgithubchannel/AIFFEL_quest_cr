###1. 텍스트 전처리 및 데이터 로드

```python
# 필요한 라이브러리 설치
!pip install pandas nltk scikit-learn textblob

# 라이브러리 임포트
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from textblob import TextBlob
from collections import Counter
from wordcloud import WordCloud
import seaborn as sns
import matplotlib.pyplot as plt
import re

# NLTK 파일 다운로드 (한 번만 실행하면 됨)
nltk.download('punkt')  # Tokenizer 필요한 파일 다운로드
nltk.download('punkt_tab') # 추가적인 Tokenization 관련 파일 다운로드
nltk.download('stopwords')  # Stopwords 목록 다운로드
nltk.download('vader_lexicon') #Vader 필요한 파일 다운로드

# 영어 불용어 목록 설정
stop_words = set(stopwords.words("english"))
```

```python
# description 컬럼에서 텍스트만 추출
descriptions = df_main['description'].dropna()

# 텍스트 클린징 함수 (불용어 제거 및 소문자 변환)
def clean_text(text):
    text = text.lower()
    words = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    cleaned_text = ' '.join([word for word in words if word.isalnum() and word not in stop_words])
    return cleaned_text

# 각 description에 대해 클린징 적용
descriptions_cleaned = descriptions.apply(clean_text)
```

###2. 주요 키워드 추출 (TF-IDF)

2-1. 상위 20개 단어 추출하여 표로 구현

```python
# TF-IDF Vectorizer 초기화
vectorizer = TfidfVectorizer(stop_words='english', max_features=20)  # 상위 20개의 키워드만 추출

# TF-IDF 행렬 생성
X = vectorizer.fit_transform(descriptions_cleaned)

# 키워드 추출
keywords = vectorizer.get_feature_names_out()
df_keywords = pd.DataFrame(X.toarray(), columns=keywords)

# 결과 출력
df_keywords
```

2-2. 상위 20개 단어 막대 그래프 시각화

```python
# 각 키워드의 TF-IDF 합 계산 (각 키워드의 빈도수로 사용)
tfidf_sum = X.sum(axis=0).A1  # 행렬의 열을 합산하여 각 키워드의 TF-IDF 합 구하기

# 데이터프레임으로 변환
df_top_words = pd.DataFrame(list(zip(keywords, tfidf_sum)), columns=["Word", "Count"])

# TF-IDF 값의 합을 기준으로 내림차순으로 정렬
df_top_words = df_top_words.sort_values(by="Count", ascending=False)

# 단어 빈도수 시각화 함수
def plot_bar_chart(df, title, color_palette):
    plt.figure(figsize=(12, 6))
    sns.barplot(x="Count", y="Word", data=df, palette=color_palette)
    plt.title(title, fontsize=14)
    plt.xlabel("TF-IDF Value")
    plt.ylabel("Keywords")
    plt.show()

# 막대그래프 시각화 실행
plot_bar_chart(df_top_words, "Top 20 Keywords (Stopwords Removed)", "viridis")
```

2-3. 전체 키워드의 빈도수를 워드 클라우드로 시각화

```python
# 모든 description 텍스트 합치기
text = ' '.join(descriptions_cleaned)

# 워드 클라우드 생성
wordcloud = WordCloud(stopwords='english', background_color='white').generate(text)

# 워드 클라우드 시각화
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

###3. 감성 분석 (긍정/부정)

3-1. 긍정/부정 감성의 이진분류

```python
# 감성 분석 함수 (Binary)
def analyze_sentiment(text):
    analysis = TextBlob(text)
    # 감성 점수: 0이 부정, 1이 긍정
    return 'positive' if analysis.sentiment.polarity > 0 else 'negative'

# 각 description에 대해 감성 분석 적용
sentiment = descriptions_cleaned.apply(analyze_sentiment)

# 결과 출력
df_main['sentiment'] = sentiment
df_main[['description', 'sentiment']]
```

3-2. 긍정/부정 감성의 이진분류를 파이차트로 시각화

```python
# 긍정/부정 감성 분포 계산
sentiment_counts = df_main['sentiment'].value_counts()

# 파이차트 시각화
plt.figure(figsize=(8, 8))
sentiment_counts.plot.pie(autopct='%1.1f%%', colors=["lightblue", "lightcoral"], startangle=90, explode=(0.1, 0))
plt.title('Sentiment Distribution (Positive vs Negative)', fontsize=14)
plt.ylabel('')  # y축 라벨 숨기기
plt.show()
```

3-3. 긍정/부정 감성의 정도를 연속적인 값으로 분석

```python
# VADER 감성 분석기 초기화
sid = SentimentIntensityAnalyzer()

# 감성 분석 함수
def vader_sentiment(text):
    sentiment = sid.polarity_scores(text)
    return sentiment['compound']  # 긍정/부정 점수

# 각 description에 대해 감성 분석 적용
descriptions_sentiment_vader = descriptions_cleaned.apply(vader_sentiment)

# 감성 결과 확인
descriptions_sentiment_vader
```

3-4. 긍정/부정 감성점수를 히스토그램으로 시각화 (감성 점수가 -1~1 범위 내에서 어떻게 분포하는지 확인)

```python
# 감성 점수의 분포 시각화
plt.figure(figsize=(10, 6))
sns.histplot(descriptions_sentiment_vader, kde=True, color='blue')
plt.title('Sentiment Score Distribution')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.show()
```

3-5. 긍정/중립/부정 감성점수를 막대그래프로 시각화

```python
plt.figure(figsize=(8, 5))

# 감성 점수를 카테고리(긍정/부정/중립)로 변환
sentiment_labels = descriptions_sentiment_vader.apply(lambda x: "Positive" if x > 0 else ("Negative" if x < 0 else "Neutral"))

# 시각화
sns.countplot(x=sentiment_labels, palette={"Positive": "green", "Negative": "red", "Neutral": "gray"})
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment Category")
plt.ylabel("Count")
plt.show()
```

3-6. 긍정/중립/부정 감성점수를 박스플롯으로 시각화

```python
# 박스플롯 시각화
plt.figure(figsize=(8, 5))
sns.boxplot(x=sentiment_labels, y=descriptions_sentiment_vader, palette={"Positive": "green", "Negative": "red", "Neutral": "gray"})
plt.title("Sentiment Score Distribution by Category")
plt.xlabel("Sentiment Category")
plt.ylabel("Sentiment Score")
plt.show()
```

3-7. 감성 점수와 콘텐츠 길이의 관계 (산점도): 길이가 긴 콘텐츠일수록 감성이 어떻게 변화하는지 확인

```python
# 산점도 그리기
df_main["description_length"] = df_main["description"].apply(lambda x: len(str(x)))

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_main["description_length"], y=descriptions_sentiment_vader, alpha=0.5)
plt.title("Sentiment Score vs. Description Length")
plt.xlabel("Description Length")
plt.ylabel("Sentiment Score")
plt.show()
```

###4. 긍정/부정 감성에 따른 주요 키워드 분석

4-1. 긍정/부정 감성에 따른 상위 20개 단어 각각 추출하여 표로 구현

```python
# 기본 불용어 리스트 + 추가적인 불용어
stop_words = set(stopwords.words('english'))
custom_stopwords = set(["movie", "film", "series", "watch", "story", "one", "two", "show", "many"])
all_stopwords = stop_words.union(custom_stopwords)

# 불용어 제거 함수
def remove_stopwords(text):
    words = word_tokenize(text)  # 단어 토큰화
    filtered_words = [word for word in words if word.lower() not in all_stopwords]  # 불용어 제거
    return " ".join(filtered_words)

# 긍정적인 감성을 가진 콘텐츠만 필터링 후 불용어 제거
positive_text = " ".join(
    df_main[descriptions_sentiment_vader > 0]["description"].dropna().apply(remove_stopwords)
)

# 부정적인 감성을 가진 콘텐츠만 필터링 후 불용어 제거
negative_text = " ".join(
    df_main[descriptions_sentiment_vader < 0]["description"].dropna().apply(remove_stopwords)
)

# 불용어 및 특수문자 제거 적용
positive_text_clean = clean_text(positive_text)
negative_text_clean = clean_text(negative_text)

# 단어 빈도수 계산 함수
def get_word_frequencies(text):
    words = text.split()  # 공백 기준으로 단어 나누기
    word_counts = Counter(words)  # 단어별 등장 횟수 계산
    return word_counts

# 불용어 제거된 텍스트 사용
positive_word_counts = get_word_frequencies(positive_text_clean)
negative_word_counts = get_word_frequencies(negative_text_clean)

# 상위 20개 단어만 선택
top_positive_words = dict(positive_word_counts.most_common(20))
top_negative_words = dict(negative_word_counts.most_common(20))

# 데이터프레임 변환
df_positive = pd.DataFrame(top_positive_words.items(), columns=["Word", "Count"])
df_negative = pd.DataFrame(top_negative_words.items(), columns=["Word", "Count"])

# 출력 확인
print("\nTop 20 Positive Keywords:")
df_positive
```

```python
print("\nTop 20 Negative Keywords:")
df_negative
```

4-2. 긍정/부정 감성에 따른 상위 20개 단어 각각 막대 그래프 시각화

```python
# 막대 그래프 시각화 함수 (단어 빈도수 기반)
def plot_bar_chart(df, title, color_palette):
    # 데이터프레임에서 'Word'와 'Count' 컬럼을 사용
    plt.figure(figsize=(12, 6))
    sns.barplot(x="Count", y="Word", data=df, palette=color_palette)  # "Count" 사용
    plt.title(title, fontsize=14)
    plt.xlabel("Word Frequency")
    plt.ylabel("Keywords")
    plt.show()

# 긍정적 텍스트의 상위 20개 단어 시각화
plot_bar_chart(df_positive, "Top 20 Positive Keywords (Stopwords Removed)", "Blues_r")

# 부정적 텍스트의 상위 20개 단어 시각화
plot_bar_chart(df_negative, "Top 20 Negative Keywords (Stopwords Removed)", "Reds_r")
```

4-3. 전체 키워드의 긍정/부정 감성의 단어 빈도수를 워드 클라우드로 시각화

```python
# 텍스트 전처리 함수 (알파벳과 공백만 남기기)
def clean_text(text):
    return re.sub(r'[^a-zA-Z\s]', '', text)

# 불용어 제거 함수
def remove_stopwords(text):
    # 텍스트 정리 (불필요한 문자는 모두 제거)
    clean = clean_text(text)

    # 단어 토큰화
    words = word_tokenize(clean)

    # 불용어 제거 후 길이가 1인 단어는 제외
    filtered_words = [word for word in words if word.lower() not in all_stopwords and len(word) > 1]

    return " ".join(filtered_words)

# 긍정적인 감성을 가진 콘텐츠만 필터링 후 불용어 제거
positive_text = " ".join(
    df_main[descriptions_sentiment_vader > 0]["description"].dropna().apply(remove_stopwords)
)

# 부정적인 감성을 가진 콘텐츠만 필터링 후 불용어 제거
negative_text = " ".join(
    df_main[descriptions_sentiment_vader < 0]["description"].dropna().apply(remove_stopwords)
)

# 워드클라우드 시각화 함수
def plot_wordcloud(text, title):
    wordcloud = WordCloud(
        width=800, height=400, background_color="white", colormap="coolwarm"
    ).generate(text)

    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.title(title, fontsize=14)
    plt.axis("off")
    plt.show()

# 긍정적 콘텐츠의 주요 키워드
plot_wordcloud(positive_text, "Positive Content Keywords")

# 부정적 콘텐츠의 주요 키워드
plot_wordcloud(negative_text, "Negative Content Keywords")
```

4-4. 긍정/부정 감성 각각이 고유하게 가지고 있는 단어들을 빈도수를 바탕으로 워드 클라우드로 시각화

```python
# 함수 정의
def map_ratings(df, mapping_dict):
    """
    데이터프레임의 'rating' 열에 있는 값들을 mapping_dict에 따라 새로운 값으로 변경합니다.

    Parameters:
    df (pandas.DataFrame): 'rating' 열이 있는 데이터프레임
    mapping_dict (dict): 원래 등급을 새 등급으로 매핑하는 딕셔너리

    Returns:
    pandas.DataFrame: 등급이 변경된 데이터프레임 사본
    """
    # 데이터프레임의 사본을 만들어 원본을 변경하지 않음
    df_copy = df.copy()

    # 매핑 적용
    df_copy['rating'] = df_copy['rating'].map(mapping_dict).fillna(df_copy['rating'])

    return df_copy

# 영화 등급 빈도 계산 및 정렬
def analyze_movie_ratings(df):
    """
    영화 등급별 빈도를 계산하고 내림차순으로 정렬하여 반환합니다.

    Parameters:
    df (pandas.DataFrame): 영화 정보가 포함된 데이터프레임, 'rating' 열이 있어야 함

    Returns:
    pandas.DataFrame: 등급별 빈도가 포함된 데이터프레임
    """
    # 등급별 영화 수 계산
    rating_counts = df['rating'].value_counts().reset_index()
    rating_counts.columns = ['rating', 'count']

    # 등급별 영화 비율 계산
    total_movies = len(df)
    rating_counts['percentage'] = (rating_counts['count'] / total_movies * 100).round(2)

    return rating_counts

# 시각화 함수
def plot_top_ratings(rating_counts, top_n=10):
    """
    상위 n개 영화 등급의 빈도를 막대 그래프로 시각화합니다.

    Parameters:
    rating_counts (pandas.DataFrame): 등급별 빈도가 포함된 데이터프레임
    top_n (int): 출력할 상위 등급 수
    """
    # 상위 n개만 선택
    top_ratings = rating_counts.head(top_n)

    plt.figure(figsize=(12, 6))
    plt.bar(top_ratings['rating'], top_ratings['count'])
    plt.title(f'Top {top_n} Movie Ratings by Frequency')
    plt.xlabel('Rating')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    for i, v in enumerate(top_ratings['count']):
        plt.text(i, v + 5, str(v), ha='center')
    plt.tight_layout()
    plt.show()

# 메인 실행 코드
if __name__ == "__main__":
    # 데이터 불러오기 (예: 'df_main'이 이미 로드되어 있다고 가정)
    # df_main = pd.read_csv('movies_data.csv')

    # 1. 먼저 원본 데이터의 등급 분포 확인
    original_rating_counts = analyze_movie_ratings(df_main)
    print("원본 영화 등급 빈도 (내림차순):")
    display(original_rating_counts)  # display()를 사용하여 깔끔하게 표 출력

    # 2. 등급 매핑 - 수정된 매핑 방향 (실제 데이터에 맞게 조정 필요)
    specific_mapping = {'13+': 'PG-13', '18+': 'R', '7+': 'TV-Y7'}  # 또는 필요에 맞게 변경
    df_modified = map_ratings(df_main, specific_mapping)

    # 3. 변경된 등급의 분포 확인
    modified_rating_counts = analyze_movie_ratings(df_modified)
    print("\n변경된 등급 분포:")
    display(modified_rating_counts.head()) # 변경된 등급 데이터의 상위 5개 행 출력

    # 4. 상위 5개 등급 시각화 (수정된 데이터 사용)
    plot_top_ratings(modified_rating_counts, top_n=5)

    # 5. 가장 흔한 등급의 영화 목록 확인 (수정된 데이터 사용)
    most_common_rating = modified_rating_counts.iloc[0]['rating']
    common_movies = df_modified[df_modified['rating'] == most_common_rating].head()

    print("\n가장 흔한 등급의 영화 목록 (상위 5개):")
    display(common_movies)  # 가장 흔한 등급의 영화 5개 출력

```

# 3. 10년 전 유행하던 장르는 현재에도 여전히 인기가 있는가?

```python
# 'United States' 데이터를 필터링
# (미국에서 제작된 콘텐츠만을 분석 대상으로 선정)
df_us = df_nf[df_nf['country'] == 'United States'].copy()

# 'release_year'을 정수형으로 변환
# (데이터 분석을 위해 문자열에서 정수형으로 변경)
df_us['release_year'] = df_us['release_year'].astype(int)

# 장르 데이터를 리스트 형태로 변환
# (각 콘텐츠는 여러 개의 장르를 가질 수 있으므로 리스트로 변환 후 확장)
df_us['listed_in'] = df_us['listed_in'].str.split(', ')
df_exploded = df_us.explode('listed_in')

# 연도별 장르 트렌드 분석
# (연도별로 각 장르의 등장 횟수를 집계하고, NaN 값은 0으로 대체)
genre_trend = df_exploded.groupby(['release_year', 'listed_in']).size().unstack().fillna(0)

# 최근 10년간 장르 트렌드 시각화
plt.figure(figsize=(12, 6))
genre_trend.tail(10).T.plot(kind='bar', stacked=True, figsize=(15, 7))
plt.title('Trends of Content Genres in the US (Last 10 Years)')
plt.xlabel('Genre')
plt.ylabel('Number of Contents')
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# 워드클라우드 생성
# (가장 많이 등장하는 장르를 한눈에 볼 수 있도록 시각화)
text = ' '.join(df_exploded['listed_in'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='black').generate(text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of US Content Genres')
plt.show()

# 장르별 콘텐츠 수 변화 분석
# (가장 인기 있는 상위 10개 장르를 선정하고 시각화)
genre_count = df_exploded['listed_in'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_count.index, y=genre_count.values, palette='viridis')
plt.xticks(rotation=45)
plt.title('Top 10 Popular Content Genres in the US')
plt.xlabel('Genre')
plt.ylabel('Number of Contents')
plt.show()
```
