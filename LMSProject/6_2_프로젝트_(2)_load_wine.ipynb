```python
import pandas as pd

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import SGDClassifier, LogisticRegression
```

```python
wine = load_wine()
wine_data = wine.data
wine_label = wine.target

print(wine.target_names)
df = pd.DataFrame(wine_data, columns=[i for i in range(wine_data.shape[1])])
df.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(wine_data, wine_label, test_size=0.2, random_state=7)
```

```python
dt_model = DecisionTreeClassifier(random_state=7)
dt_model.fit(X_train, y_train)
dt_pred = dt_model.predict(X_test)

rf_model = RandomForestClassifier(random_state=7)
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)

svm_model = SVC()
svm_model.fit(X_train, y_train)
svm_pred = svm_model.predict(X_test)

sgd_model = SGDClassifier(random_state=7)
sgd_model.fit(X_train, y_train)
sgd_pred = sgd_model.predict(X_test)

lr_model = LogisticRegression(max_iter=1000, random_state=7)
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)
```

```python
# (6) 모델을 평가해 보기
# 다중 클래스 분류이며, 각 클래스에 동일한 중요도를 부여 -> Precision/Recall, F1-Score가 적절한 평가 지표

models = {
    "decision tree": dt_pred,
    "random forest": rf_pred,
    "svm": svm_pred,
    "sgd": sgd_pred,
    "logistic regression": lr_pred
}

for name, y_pred in models.items():
  print(f"\n[{name}]")
  print(classification_report(y_test, y_pred))
```
